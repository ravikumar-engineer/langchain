### Hugging Face Sentence Transformers



```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2") # all-mpnet-base-v2 (768 dim, high quality)
```

```python
text1 = "Alpha is the first letter of the Greek alphabet."
query_result = embeddings.embed_query(text1)
query_result
```

```python
len(query_result)
```

```python
text2 = "Beta is the second letter of the Greek alphabet."
doc_result = embeddings.embed_documents([text1, text2])
doc_result[0] #, doc_result[1]
```
